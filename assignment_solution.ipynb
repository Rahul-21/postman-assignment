```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.ExcelFile('sales_data.xlsx')
```

```python
df2=pd.ExcelFile('date.xlsx')
```

```python
x1 = df.parse()
x2 = df2.parse()
x1.head()
x1.info()
```

```python
x2.head()
```

```python
x1=pd.DataFrame(data=x1)
x2=pd.DataFrame(data=x2)
```

```python
x3= pd.concat([x1, x2], axis=1, join='inner')
x3.info()
```

```python
x3.drop(['transaction timestamp'],axis=1)
x3.head()
```

```python
x3.columns
```

```python
x3=pd.DataFrame(data=x3)
```

```python
x3["Total Price"] = x3["quantity sold"] * x3["unit price"]
```

```python
x3.head()
```

```python
100*(x3.isnull().sum()/len(x3.index))
```

```python
x3.info()
```

```python
x3.groupby("month_of_year").sum().sort_values("Total Price", ascending=False).head()
```

```python
group = x3.groupby(["transaction country","month_of_year"]).sum()
total_price = group["Total Price"].groupby(level=0, group_keys=False)
total_price.nlargest(5)
```

```python
group = x3.groupby(["customer id","product description"]).sum()
total_price = group["Total Price"].groupby(level=0, group_keys=False)
total_price.nlargest(5)
```

```python
x3.sort_values("Total Price")
```

```python
x3['condition']=x3['Total Price'].apply(lambda x:1 if x>0 else 0)
x3['condition']=x3['condition'].apply(lambda x:pd.to_numeric(x))
x3['condition'].value_counts()
```

```python
#group=x3.groupby(['month_of_year'])
plt.figure(figsize=(10, 4))
sns.countplot(x='month_of_year',hue="condition" ,data=x3)
plt.show()
```

```python
x3['date']=pd.to_datetime(x3['transaction timestamp'],infer_datetime_format=True)
x3=x3.drop('transaction timestamp',axis=1)
x3.info()
x3.head()
```

```python
x3['date'] = x3['date'].dt.year
x3.head()
```

```python
plt.figure(figsize=(10, 4))
sns.countplot(x='date',data=x3)
plt.show()
```

```python
plt.figure(figsize=(25, 5))
sns.countplot(x='day_name',data=x3)
plt.show()
```
